```python
import cv2
import numpy as np

i1 = cv2.imread("inter1.png")
i2 = cv2.imread("inter2.png")

i1 = cv2.resize(i1, (400,400))
i2 = cv2.resize(i2, (400,400))

# cv2.imshow("imagen",i1)
# cv2.imshow("im2",i2)

i = cv2.cvtColor( i2-i1 , cv2.COLOR_BGR2GRAY)

b = np.array( np.where( i==0 , 255 , 0   ) , dtype=np.uint8)

# print(b.dtype)
cv2.imshow("i",i)
cv2.imshow("b",b)


alpha = 0.7
blend = np.array(alpha*i1 + (1-alpha)*i2,dtype=np.uint8)

cv2.imshow("blend",blend)


#####
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
i1_r = i1.reshape( (-1,3))
# print(i1_r)
km.fit(i1_r)

print(km.cluster_centers_)
print(km.labels_)
#######

i1_g = cv2.cvtColor( i1 , cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(i1_g,140,255,0)
cv2.imshow("thresh",thresh)
contours,hierarchy = cv2.findContours(thresh, 1, 2)

x,y,w,h = cv2.boundingRect(contours[0])
print("i1 - bbb",x,y,w,h)
    
for cnt in contours:
    
#     cnt = contours[0]
#     print(cnt)

    x,y,w,h = cv2.boundingRect(cnt)
#     print("i1",x,y,w,h)
    cv2.rectangle(i1,(x,y),(x+w,y+h),(0,255,0),2)

    
    cv2.imshow("bound",i1)

print("_______________-")


i1_g = cv2.cvtColor( i2 , cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(i1_g,140,255,0)
cv2.imshow("thresh",thresh)
contours,hierarchy = cv2.findContours(thresh, 1, 2)
x,y,w,h = cv2.boundingRect(contours[0])
print("i2 - bbb ",x,y,w,h)
for cnt in contours:
    
#     cnt = contours[0]
#     print(cnt)

    x,y,w,h = cv2.boundingRect(cnt)
#     print("i2",x,y,w,h)
    cv2.rectangle(i2,(x,y),(x+w,y+h),(0,255,0),2)

    
    cv2.imshow("bound2",i2)





cv2.waitKey(0)
cv2.destroyAllWindows()
```
